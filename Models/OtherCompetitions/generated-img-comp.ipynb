```python
import ultralytics
import torch
import os
import json
import yaml
import pandas as pd
import shutil
```

```python
os.listdir('/kaggle/input')
```

```python
train = pd.read_csv('/kaggle/input/image-classification-real-or-ai-generated-photo/train.csv')
test = pd.read_csv('/kaggle/input/image-classification-real-or-ai-generated-photo/test.csv')
```

```python
os.mkdir('/kaggle/working/dataset')
os.mkdir('/kaggle/working/dataset/train')
os.mkdir('/kaggle/working/dataset/test')


os.mkdir(f'/kaggle/working/dataset/train/0')
os.mkdir(f'/kaggle/working/dataset/test/0')

os.mkdir(f'/kaggle/working/dataset/train/1')
os.mkdir(f'/kaggle/working/dataset/test/1')
```

```python
fromDir = '/kaggle/input/image-classification-real-or-ai-generated-photo/train/train/'

for row in train.itertuples(): # idx1=filename, idx2=class
    shutil.copyfile(fromDir + row[1], f'/kaggle/working/dataset/train/{row[2]}/{row[1]}')
    shutil.copyfile(fromDir + row[1], f'/kaggle/working/dataset/test/{row[2]}/{row[1]}')
```

```python
if torch.cuda.is_available():
    print("GPU is available")
    device = torch.device("cuda")
else:
    print("GPU is not available, using CPU instead")
    device = torch.device("cpu")
```

```python
model = ultralytics.YOLO('yolov8m-cls.pt')

model.to(device)
```

```python
# Currently not using wandb
os.environ['WANDB_DISABLED'] = 'true'
```

```python
model.train(data='/kaggle/working/dataset', epochs=5, imgsz=224)
```

```python
os.mkdir('/kaggle/working/predimg')
```

```python
fromDir = '/kaggle/input/image-classification-real-or-ai-generated-photo/test/test/'
for img in os.listdir(fromDir):
    shutil.copy(fromDir + img, f'/kaggle/working/predimg/{img}')
```

```python
ids = []
classpred = []

for row in test.itertuples():
    ids.append(row[1])
    filename = row[1]
    prediction = model(f'/kaggle/working/predimg/{filename}')
    classpred.append(prediction[0].names[int(prediction[0].probs.data.argmax().cpu().numpy())])
```

```python
pred_df = pd.DataFrame({'Image': ids, 'Label': classpred})
```

```python
pred_df.to_csv('gen_img_submission.csv', index=False)
```
